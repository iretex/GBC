```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# # of hours a AI student spends on their courses in a month
np.random.seed(10)
no_hours = np.random.normal(40.0, 10.0, 10000)

plt.hist(no_hours, 50)
plt.show()
```

```python
# Calculate and report the mean, std, and var of the data shown above.
# What is the theoretical relationship between std and var? Do the numbers reperted in the previous step agree with your expectation?
# Begin changing the different parameters on the normal function, discuss your observations as you change each input parameter.
```

```python
mean = np.mean(no_hours)
std = np.std(no_hours)
var = np.var(no_hours)

print(f"The mean of the data is: {mean}")
print(f"The std of the data is: {std}")
print(f"The var of the data is: {var}")
```

The theoretical relationship between the std and var is $std = \sqrt var$

Since $9.896^2 = 97.943$, the theoretical relaationship holds true for the data.

```python
no_hours = np.random.normal(60.0, 10.0, 10000)

plt.hist(no_hours, 50)
plt.show()

mean = np.mean(no_hours)
std = np.std(no_hours)
var = np.var(no_hours)

print(f"The mean of the data is: {mean}")
print(f"The std of the data is: {std}")
print(f"The var of the data is: {var}")
```

**Observation**: 
In this instance, I have increated the mean from 40 to 60 and the curve is now centered around the new mean value of 60

```python
no_hours = np.random.normal(40.0, 30.0, 10000)

plt.hist(no_hours, 50)
plt.show()

mean = np.mean(no_hours)
std = np.std(no_hours)
var = np.var(no_hours)

print(f"The mean of the data is: {mean}")
print(f"The std of the data is: {std}")
print(f"The var of the data is: {var}")
```
